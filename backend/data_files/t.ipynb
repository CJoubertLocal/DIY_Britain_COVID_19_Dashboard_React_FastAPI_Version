```python
import pandas as pd
import json
```

```python
eng = None
with open('./england_vaccination_percentages.json', "rt") as INFILE:
    eng = json.load(INFILE)

scot = None
with open('./scotland_vaccination_percentages.json', "rt") as INFILE:
    scot = json.load(INFILE)
```

```python
e = pd.DataFrame.from_dict(eng['data'])
e['date'] = pd.to_datetime(e['date'])
s = pd.DataFrame.from_dict(scot['data'])
s['date'] = pd.to_datetime(s['date'])
```

```python
r = pd.merge(e, s, how='outer', on='date')
r = r.sort_values(by=['date'], ascending=True)
r.to_json('./r.json')
```

```python
r = r.replace('null', pd.NA).ffill()
r.to_json('./r2.json')
```
